```python
import librosa
import torch
import IPython.display as display
# fine tune
from transformers import Wav2Vec2ForCTC, AutoProcessor
import numpy as np
```

```python
processor = AutoProcessor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

Once your audio file is uploaded (e.g., named `audio.wav` and located in `/content/`), you can load and process it using the following steps:

```python
# Define the path to your uploaded audio file
AUDIO_FILE_PATH = "/content/harvard.wav" # Updated to use the provided file

# Load the audio file
# Wav2Vec2 models typically expect a sample rate of 16kHz
try:
    speech, rate = librosa.load(AUDIO_FILE_PATH, sr=16000)
    print(f"Audio loaded: Sample rate = {rate} Hz, Duration = {len(speech)/rate:.2f} seconds")

    # Display the audio (optional)
    display.Audio(speech, rate=rate)

except FileNotFoundError:
    print(f"Error: Audio file not found at {AUDIO_FILE_PATH}. Please ensure it is uploaded correctly.")
except Exception as e:
    print(f"An error occurred while loading the audio file: {e}")
```

Now, let's tokenize the audio and pass it through the model to get predictions.

```python
# Tokenize the input speech using the processor
input_values = processor(speech, sampling_rate=rate, return_tensors="pt").input_values

# Get logits from the model
with torch.no_grad():
    logits = model(input_values).logits

# Take argmax to get the predicted token IDs
predicted_ids = torch.argmax(logits, dim=-1)

# Decode the predicted IDs to text using the processor
transcription = processor.batch_decode(predicted_ids)

print("Transcription:")
print(transcription[0])
```
